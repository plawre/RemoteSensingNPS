```python
# This downloads all WOA13 variables at 1 degree resolution. It works!
import os
import requests
from bs4 import BeautifulSoup

# Base URL of the NOAA data directory
base_url = 'https://www.ncei.noaa.gov/data/oceans/woa/WOA13/DATAv2/'

# Directory to save the downloaded files
base_save_dir = '/expanse/lustre/projects/unc107/yli39/data/WOA13/DATAv2/'

# Ensure the base save directory exists
if not os.path.exists(base_save_dir):
    os.makedirs(base_save_dir)

# Function to download a file from a given URL and save to the specified directory
def download_file(file_url, save_dir):
    local_filename = os.path.join(save_dir, file_url.split('/')[-1])
    with requests.get(file_url, stream=True) as r:
        r.raise_for_status()
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
    return local_filename

# Get the list of subdirectories and files from the base URL
response = requests.get(base_url)
soup = BeautifulSoup(response.content, 'html.parser')

# List of variables to download
variables = ['AOU', 'conductivity', 'density', 'nitrate', 'o2sat', 'oxygen', 'phosphate', 'salinity', 'sea_water_electrical_conductivity', 'sea_water_sigma', 'sea_water_temperature', 'silicate', 'temperature']

# Iterate over each variable directory
for var in variables[0:1]:
    var_dir_url = base_url + var + '/netcdf/all/1.00/'
    save_dir = os.path.join(base_save_dir, var, 'netcdf', 'all', '1.00')
    
    # Ensure the save directory for the variable exists
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    # Get the list of files in the variable directory
    response = requests.get(var_dir_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    file_links = soup.find_all('a')
    
    # Download each file in the variable directory
    for file_link in file_links:
        if not file_link.get('href').endswith('/'):
            file_url = var_dir_url + file_link.get('href')
            print(f'Downloading {file_url} to {save_dir}')
            download_file(file_url, save_dir)
```
